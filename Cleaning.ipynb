```python
df1 = df['Ship Mode'].mode()[0] #display the most frequent value
df1
```

```python
df.fillna(df1, inplace=True) #fill the null values with most frequent value
df
```

```python
df.rename(columns={'Order Id': 'order_id', 'Order Date': 'order_date', 'Ship Mode': 'ship_mode', 'Segment': 'segment', 'Country':'country', 'City': 'city', 'State': 'state', 'Postal Code': 'postal_code', 'Region': 'region', 'Category': 'category', 'Sub Category': 'sub_category', 'Product Id': 'product_id', 'cost price':'cost_price', 'List Price': 'list_price', 'Quantity': 'quantity','Discount Percent': 'discount_percent'}, inplace=True) #standardising column names
```

```python
df
```

```python
df["order_id"] = df["order_id"].astype(int)                
df["order_date"] = pd.to_datetime(df["order_date"])        
df["ship_mode"] = df["ship_mode"].astype(str)
df["segment"] = df["segment"].astype(str)
df["country"] = df["country"].astype(str)
df["city"] = df["city"].astype(str)
df["state"] = df["state"].astype(str)
df["postal_code"] = df["postal_code"].astype(int) 
df["region"] = df["region"].astype(str)
df["category"] = df["category"].astype(str)
df["sub_category"] = df["sub_category"].astype(str)
df["product_id"] = df["product_id"].astype(str)
df["cost_price"] = df["cost_price"].astype(int)
df["list_price"] = df["list_price"].astype(int)
df["quantity"] = df["quantity"].astype(int)
df["discount_percent"] = df["discount_percent"].astype(int)


```

```python
df['discount'] = df.apply(lambda row: (row['discount_percent'] * row['list_price'])/100, axis=1) #create a new column "discount"
df["discount"] = df["discount"].astype(int)
df
```

```python
df['sales_price'] = df.apply(lambda row: row['list_price'] - (row['list_price'] * row['discount_percent'] / 100), axis=1) #create a new column "sales_price"
df["sales_price"] = df["sales_price"].astype(int)
df
```

```python
df['profit'] = df.apply(lambda row: row['sales_price'] - row['cost_price'], axis=1) #create a new column "profit"
df["profit"] = df["profit"].astype(int)
df
```

```python
df['revenue'] = df.apply(lambda row: row['sales_price'] * row['quantity'], axis=1) #claculate revenue
df["revenue"] = df["revenue"].astype(int)
df
```
